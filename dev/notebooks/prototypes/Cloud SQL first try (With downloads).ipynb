This notebook attempts to install Cloud SQL Proxy from notebook cell.

This works but installing from Dockerfile is preferred.

## Install Cloud SQL Proxy.

```python
%%bash
wget "https://dl.google.com/cloudsql/cloud_sql_proxy.linux.amd64"
```

```python
%%bash
mv cloud_sql_proxy.linux.amd64 cloud_sql_proxy
chmod +x cloud_sql_proxy
sudo mkdir /cloudsql; chmod 777 /cloudsql
```

## Install python-mysqldb.

```python
%%bash
apt-get  --assume-yes install python-mysqldb
```

## Try starting the proxy.

```python
%%bash
./cloud_sql_proxy -dir=/cloudsql -credential_file=/content/datalab/.config/credentials -instances=datalab-deploy-test1:us-central1:cloudsql
```
